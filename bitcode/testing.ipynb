```python
import matplotlib.pyplot as plt

def graficar_puntos(secuencia_puntos:list[0]):
    plt.plot(secuencia_puntos)
    plt.xlabel('Secuencia')
    plt.ylabel('Precio')
    plt.title('Valor de Bitcoin')
    plt.grid(True)
    plt.show()
```

```python
from binance.client import Client

from dataclasses import dataclass

@dataclass
class Ticker:
    price: float = 0.0
    symbol: str = ""

    @classmethod
    def from_dict(cls, data: dict):
        return cls(
            price=data.get("price", 0.0),
            symbol=data.get("symbol", "")
        )
    
class Bot:
  def __init__(self):
    # Configura tu API Key
    api_key = ""
    api_secret = ""
    self.client = Client(api_key, api_secret)

  def getTicker(self, symbol):
    ticker = self.client.get_symbol_ticker(symbol=symbol)
    return Ticker.from_dict(data=ticker)
```

```python
import time
bot = Bot()

valores = []
for i in range(1, 11):
    response = bot.getTicker("BTCUSDT")
    precio = float(response.price)
    valores.append(round(precio, 2))
    print (f"{i}) {response}")
    time.sleep(5)

graficar_puntos(valores)
```
